```python
def check_input(item):
    return isinstance(item, str)

def get_substring(s):
    max_sub = ""
    for i in range(len(s)):
        current_sub = ""
        for char in s[i:]:
            if char not in current_sub:
                current_sub += char
            else:
                break
        if len(current_sub) > len(max_sub):
            max_sub = current_sub
    return max_sub

def main(data_list):
    longest_str = ""
    target_index = -1
    for index, item in enumerate(data_list):
        if check_input(item):
            if len(item) > len(longest_str):
                longest_str = item
                target_index = index
    if target_index == -1:
        print(-1, "")
    else:
        result_sub = get_substring(longest_str)
        print(target_index, f"'{result_sub}'")

main([[0, 4, 8], 'abca', 7, 'aaaaa'])
```
